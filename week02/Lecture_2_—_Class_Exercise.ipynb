```python
import pandas as pd
import numpy as np

# World Happiness Report 2023 — representative data
# Source: https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023

df = pd.read_csv('world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

Task 1 — Regional Comparison Bar Chart

```python
region_avg = (df.groupby('Region')['Happiness_Score']
              .mean()
              .reset_index()
              .sort_values('Happiness_Score'))  # sort for horizontal bar

print(region_avg)

fig1 = px.bar(
    region_avg,
    x='Happiness_Score',
    y='Region',
    orientation='h',
    text=region_avg['Happiness_Score'].round(2),
    color='Happiness_Score',
    color_continuous_scale='Blues'
)

fig1.update_layout(
    title="Western Europe leads global happiness, highlighting regional inequality",
    xaxis_title="Average Happiness Score",
    yaxis_title="Region",
    template='simple_white'
)

# Zero baseline (important rule)
fig1.update_xaxes(range=[0, region_avg['Happiness_Score'].max() + 1])

fig1.show()
```

Task 2 — Bottom vs. Top: A Contrast Story

```python
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score')
global_avg = df['Happiness_Score'].mean()
print(f"Global average: {global_avg:.2f}")

fig2 = px.bar(
    combined,
    x='Happiness_Score',
    y='Country',
    orientation='h',
    color='Group',
    text=combined['Happiness_Score'].round(2),
    color_discrete_map={
        'Top 8': 'green',
        'Bottom 8': 'red'
    }
)

fig2.update_layout(
    title="Massive happiness gap between top and bottom countries reveals global inequality",
    xaxis_title="Happiness Score",
    yaxis_title="Country",
    template='simple_white'
)

# Zero baseline
fig2.update_xaxes(range=[0, combined['Happiness_Score'].max() + 1])

# Add global average reference line
fig2.add_vline(
    x=global_avg,
    line_dash="dash",
    line_color="black",
    annotation_text=f"Global Avg: {global_avg:.2f}",
    annotation_position="top"
)

fig2.show()
```
